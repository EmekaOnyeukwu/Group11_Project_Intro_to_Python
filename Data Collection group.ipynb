```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

base_url = 'http://books.toscrape.com/catalogue/category/books/nonfiction_13/index.html'
url = base_url
page_number = 1
titles = []
prices = []
availabilities = []

while True:
    response = requests.get(url)

    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        books = soup.find_all('article', class_='product_pod')

        for book in books:
            titles.append(book.h3.a['title'])
            prices.append(book.find('p', class_='price_color').get_text())
            availabilities.append(book.find('p', class_='instock availability').get_text().strip())

        next_page = soup.find('li', class_='next')

        if next_page:
            page_number += 1
            url = f"{base_url.replace('index.html', '')}page-{page_number}.html"
        else:
            break
    else:
        print('Failed to retrieve the page.')
        break
data = {
    'Title': titles,
    'Price': prices,
    'Availability': availabilities
}
df = pd.DataFrame(data)
df.to_csv('non_fiction_books.csv', index=False)
```
